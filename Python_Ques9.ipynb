```python
#Ans-9

import pandas as pd

data=pd.read_csv('C://Users//A146363//Downloads//Electric_Vehicle_Population_Data.csv')
```

1.Get all the cars and their types that do not qualify for clean alternative fuel vehicle

```python
df1=data[['VIN (1-10)','Electric Vehicle Type','Clean Alternative Fuel Vehicle (CAFV) Eligibility']][data['Clean Alternative Fuel Vehicle (CAFV) Eligibility']=='Not eligible due to low battery range'].drop_duplicates()

df1
##cars and their types that do not qualify for clean alternative fuel vehicle
```

2. Get all TESLA cars with the model year, and model type made in Bothell City.

```python
data[['VIN (1-10)','Make','Model Year','Electric Vehicle Type','City']][(data['Make']=='TESLA') & (data['City']=='Bothell')]
```

3. Get all the cars that have an electric range of more than 100, and were made after 2015

```python
data[['VIN (1-10)','Electric Range','Model Year']][(data['Electric Range']>100) & (data['Model Year']>2015)]
```

4. Draw plots to show the distribution between city and electric vehicle type

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 10))
df_city_ev = data.groupby(['City', 'Electric Vehicle Type']).size().unstack()
df_city_ev.plot(kind='barh', stacked=True)
plt.title('Distribution of Electric Vehicle Types by City')
plt.xlabel('Frequency')
plt.ylabel('City')
plt.legend(title='Electric Vehicle Type')
plt.show()
```
